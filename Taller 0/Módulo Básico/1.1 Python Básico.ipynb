# Tipos de variables
Python cuenta con los tipos de variables clásicos en programación. Cada uno de estos soporta métodos y funciones diferentes, y tiene características diferentes.

Para verificar el tipo de una variable se utiliza la función ```type()```

````python
type(2)
````

### Números
Acá el tamaño (en bits) de los números no influye en el tipo de variable que se tiene. La diferenciación más importante es entre enteros (int) y decimales (float)

````python
entero = 2
decimal = 2.0

print(type(entero))
print(type(decimal))
````

````python
entero_grande = 22222222222222222222222222222222222222
decimal_grande = 22222222222222222222222222222222222222.0

print(type(entero_grande))
print(type(decimal_grande))
````

### Strings
En python, cualquier cadena de carácteres se considera un string. No existe el elemento char. Los strings se pueden definir con comillas sencillas (```'```) o comillas dobles (```"```)

````python
caracter = "a"
palabra = "ala"

print(type(caracter))
print(type(palabra))
````

````python
string1 = "Hello, world"
string2 = 'Hello, world'

string1 == string2
````

El uso de comillas dobles o sencillas permite usarlas dentro de la string.

````python
frase1 = 'Miguel les preguntó: "¿Vamos a almorzar?"'
print(frase1)

frase2 = "It was Miguel's first party."
print(frase2)
````

# Operaciones básicas

### Numéricas

````python
suma = 5+2
resta = 5-2
multiplicacion = 5*2
division = 5/2
division_entera = 5//2
potencia = 5**2
modulo = 5 % 2

print("Si tenemos dos números 5 y 2, su suma es: " + str(suma) + ", su resta es: " + str(resta) +
      ", su multiplicación es: " + str(multiplicacion) + ", su division es: " + str(division) +
     ", su división entera es: " + str(division_entera) + ", 5 elevado a la 2 es: " + str(potencia) + 
     ", y 5 módulo 2 es: " + str(modulo))
````

### Operaciones en strings
En Python, algunos operadores matemáticos tienen sentido en variables de tipo string

````python
saludo = "Hola, "
nombre = "Miguel"
print(saludo + nombre)
````

````python
risa = "ja"
print(risa*10)
````

Sin embargo, solo se pueden hacer operaciones del mismo tipo de variable.

````python
nombre = "Juan "
numero = 4
#Esto está mal
print(nombre + numero)
````

````python
#Esto está bien
print(nombre + str(numero))
````

También se puede saber los substrings dentro de una cadena

````python
carrera = "Ing. Industrial"
print(carrera[5])
print(carrera[0:3])
print(carrera[-3])
````

### Operaciones comparativas

````python
var1 = 1
var2 = 2

print(var1 == var2)
print(var1 <= var2)
print(var1 >= var2)
print(var1 != var2)
````

### And, or
Clásicos y fáciles de usar

````python
print(True and True)

print(True and False)

print(False and False)

print(True or True)

print(True or False)

print(False or False)
````

### Not

````python
True and not False
````

### In
Operador para ver si un objeto está dentro de otro

````python
"a" in "Hola"
````

````python
"b" not in "Hola"
````

También se puede utilizar en arreglos

````python
numeros = [1,2,3,4,5,6]
1 in numeros
````

````python
9 in numeros
````

### Is
Operador para saber si un objeto es el mismo que otro (en términos del espacio que ocupa en memoria)

````python
var1 = [1,2,3]
var2 = [1,2,3]

var1 is var2
````

````python
var1 = [1,2,3]
var2 = var1

var1 is var2
````

````python
var1 = [1,2,3]
var2 = var1[:]

var1 is var2
````

### Operaciones de strings
Las strings en Python tienen varios métodos que se pueden utilizar.

````python
string = "        Análisis con Machine Learning               "
````

````python
#Pasar a mayúsculas
string.upper()
````

````python
#Pasar a minúsculas
string.lower()
````

````python
#Pasar minúsculas a mayúsculas y al revés
string.swapcase()
````

````python
#Elimina espacios al principio o al final del string
string.strip()
````

````python
#Divide el string en un arreglo de datos
string.split("a")
````

````python
#Busca el primer string y lo reemplaza por el segundo
string.replace(" ", "")
````

Para obtener datos a través de la consola, se utiliza el método <b> input() </b>.

````python
nombre = input("Nombre: ")
carrera = input("Carrera: ")
"{0} es estudiante de {1}".format(nombre, carrera)
````

Tambien para imprimir se puede utilizar `print()`. Por defecto, el separador entre variables es el espacio. También se le puede definir el final de la impresión

````python
print(nombre,"es estudiante de",carrera)
print(nombre,"es estudiante de",carrera, sep="-")
print(nombre,"es estudiante de",carrera, sep="-",end=".")
````
