# If

````python
#Ejemplo básico
x, y = 8, 3

if x < y:
    print(x, "es menor que", y)
elif x > y:
    print(x, "es mayor que", y)
else:
    print(x, "y", y, "son iguales")
````

### Objeto ```range()```
Los rangos son objetos aparte en Python, no son considerados listas pero son iterables. ¿Que hay de raro si lo defino hasta 5?

````python
for i in range(5):
    print(i)
````

El método consta de 3 parámetros: `range(start, stop, step)`. Sin embargo las funciones tienen valores por defecto. <br>
Si se pasa un parámetro a la función, la función asume que ese es el `stop` y deja `start = 0` y `step = 1`

````python
#¿Si se pasan dos parámetros?
var = range(8,10)
print("Rango con dos parámetros:",var)
for i in var:
    print(i)
#Y que pasa si se pone al revés
var = range(10,8)
print("Rango con dos parámetros al revés:",var)
for i in var:
    print(i)
#Con los tres parámetros
var = range(2,8,2)
print("Rango con dos tres parámetros:",var)
for i in var:
    print(i)
````

Es importante que este objeto no es una lista para Python, es un objeto específico que se llama ```range()```

### Recorrer una tupla

````python
#Crear el arreglo
import random
coordenadas = []
for i in range(10):
    x = int(random.uniform(-100,100))
    y = int(random.uniform(-100,100))
    tupla = (x,y)
    coordenadas.append(tupla)
````

````python
#Recorrer el arreglo
print (coordenadas)
for x, y in coordenadas:
    print("La coordenada x es : ", x, ". La coordenada y es: ", y)
````

### Recorrer un diccionario
Por default, Python recorre un diccionario por sus keys.

````python
heroes = {'Simbad': 'el marino', 'Coraje': 'el perro cobarde', 'Plinio': 'el viejo', 'Aquiles': 'el guerrero',
         'Batman': 'el caballero de la noche', 'Harry Potter': 'el elegido', 'Superman': 'el marciano'}

for heroe in heroes:
    print(heroe, heroes[heroe])
````

````python
#También se puede recorrer por keys y values al tiempo
for k, v in heroes.items():
    print(k,v)
````

# Sacarle "jugo" a Python
Dado un número, determine si este es primo o no.

````python
#Forma clásica y larga
n = int(input("Porfavor ingrese un número: "))
primo = True

for i in range(2,n):
    if n%i == 0:
        primo = False
        
print("Sí es primo") if primo else print("No es primo")
````

````python
#Forma corta
n = int(input("Porfavor ingrese un número: "))

primo = not any([n%i == 0 for i in range(2,n)])

print("Sí es primo") if primo else print("No es primo")
````
